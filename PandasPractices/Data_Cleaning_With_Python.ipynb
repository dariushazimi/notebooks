# Missing values
- Add a default value for missing data or use mean to fill it in
- Delete the row or column with missing data (but this should be your last option)
- Interpolate the rows
- Replace

##  How to check for missing data using Pandas
### Fales means no missing data
- df.isnull().sum() int
- df.isnull().any() bool

```python
df.isnull()
# anywhere we see True, it means we are missing data
# to make it easier to read
```

```python
df.isnull().any()
# this list the column with missing values
```

## Check if Entire dataset has missing values

```python
df.isnull().any().any()
# this will check the entire dataset to see if it contains any missing values
```

## Return rows which have missing values

```python
df[df.isnull().any(axis=1)]
```

## Find number of missing rows in the dataframe

```python
sum(df.isnull().values.any(axis=1))
```

## Columns with NAN using Integer

```python
df.isnull().sum()
```

## Total number of missing NAN

```python
df.isnull().sum().sum()
```

## Adding default values or Filling for Missing Data

```python
df.head()
```

```python
# Lets fill the NaN with zero
df_with_0 = df.fillna(0)
```

```python
df_with_0.head()
```

```python
## Fill missing values with mean
```

```python
df['DURATION']
```

```python
df['DURATION'].mean()
```

```python
df_with_mean = df.DURATION.fillna(df['DURATION'].mean())
# the df.DURATION ensures that we are ONLY filling the DURATION column with the mean of the DURATION column and NOT the entire DataFrame
# df_with_mean = df.DURATION.fillna(df['DURATION'].mean(), inplace=True)
```
