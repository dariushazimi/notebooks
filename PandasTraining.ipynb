# Colormaps
Classes of colormaps
Sequential = used for ordering, there is a change in lightness of the color
Diverging: should be used when the data is divergaing around the middle value
Qualitative: should be used to represent information when the data doesn't have any ordering and it is often misc colors
    
    

```python
fo['Sport'].value_counts().plot(kind='pie', colormap='Pastel1');
```

# When to use seaborn vs matplotlib
Matplotlib: short scripts and working in conjunction with pyplot or - simple plot types like line, bar, plot
Seaborn: When dealing with statistical data or categorical data, or requiring moe advanced plots
