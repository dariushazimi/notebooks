Question 1: In which events did Jesse owens win a medal

Question 2: Which country has won the most men's gold medals iin singles badminton over the years?
    

```python
myfile[(myfile['Medal'] =='Gold') & (myfile['Sport'] =='Badminton')]
```

```python
most_medals=myfile[(myfile['Medal'] =='Gold') & (myfile['Sport'] =='Badminton')]
most_medals['NOC'].value_counts()
```

Question 3: Which 3 countries have won the most medals in recent years (from 1984 to 2008)

```python
myfile[(myfile['Edition'] >= 1984) & (myfile['Edition']< 2008)]
```

```python
# Now we can do a value_counts()
cn = myfile[(myfile['Edition'] >= 1984)]
cn['NOC'].value_counts().head(3)
```

```python
# Another way is to do it in one line
(myfile[(myfile['Edition'] >=1984) & (myfile['Edition']< 2008)])['NOC'].value_counts()
```

Quesetion 4; Display the male gold meal winners for the 100m track and filed sprint event over the years.


```python
myfile[(myfile['Gender']=='Men') & (myfile['Medal'] == 'Gold') & (myfile['Event']=='100m')]
```

```python
gmh = myfile[(myfile['Gender']=='Men') & (myfile['Medal'] == 'Gold') & (myfile['Event']=='100m')]
gmh.sort_values('Edition', ascending = False)
# to only disply the columns we want we can add them as lists

gmh.sort_values('Edition', ascending = False)[['City', 'Edition', 'Athlete', 'Medal']]
```
